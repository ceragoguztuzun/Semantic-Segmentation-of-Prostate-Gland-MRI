```python
def convert_mask(path):
    images = []

    im = pydicom.dcmread(path)

    im = im.pixel_array.astype(np.uint8)
    im = im*255

    print(im.shape)

    for i in range(im.shape[0]):
        final_image = Image.fromarray(im[i])
        images.append(final_image)
    
    return images
```

```python
def get_names(path):
    fn = []
    for root, dirnames, filenames in os.walk(path):
        for filename in filenames:
            _, ext = os.path.splitext(filename)
            if ext in ['.dcm']:
                path = root+'/'+filename
                path = path.replace("\\", "/" )
                fn.append(path)
    
    return fn
```
