```python
def getLosses(fn, norm_flag=0):
    with open(fn) as f:
        lines = f.readlines()
    #    
    epochs = []
    e = 1
    if norm_flag == 1:
        lines_ = []
        lines_.append(lines[3])
        for i in range(len(lines)):
            if 'Got' in lines[i] and i > 1:
                lines_.append(lines[i-1])
                
                e += 1
        lines = lines_
    #
    losses = [float(line.split('\n')[0].split('Loss: ')[1]) for line in lines if 'Loss' in line]
    
        
    return losses, 1
```

```python

from operator import add

def getPrecisionRecall(fn, norm_flag):
    with open(fn) as f:
        lines = f.readlines()
    #
    if norm_flag == 1:    
        lines_ = []
        lines_.append(lines[2])
        for i in range(len(lines)):
            if 'Dice' in lines[i] and i > 3:
                lines_.append(lines[i-3])
        lines = lines_
    #

    TP = [float(line.split('\n')[0].split(', ')[1].split('TP: ')[1]) for line in lines if 'Tversky' in line]
    FP = [float(line.split('\n')[0].split(', ')[2].split('FP: ')[1]) for line in lines if 'Tversky' in line]
    FN = [float(line.split('\n')[0].split(', ')[3].split('FN: ')[1]) for line in lines if 'Tversky' in line]
    
    #prec = TP/TP+FP
    denom = list( map(add, TP, FP) )
    prec = [i / j for i, j in zip(TP, denom)]
    
    #recall = TP/TP+FN
    denom_ = list( map(add, TP, FN) )
    recall = [i / j for i, j in zip(TP, denom_)]

    return prec, recall
```

```python
from operator import add

def getF1(fn, norm_flag):
    with open(fn) as f:
        lines = f.readlines()
    #    
    if norm_flag == 1:
        lines_ = []
        lines_.append(lines[2])
        for i in range(len(lines)):
            if 'Got' in lines[i] and i > 2:
                lines_.append(lines[i-2])
        lines = lines_
    #

    TP = [float(line.split('\n')[0].split(', ')[1].split('TP: ')[1]) for line in lines if 'Tversky' in line]
    FP = [float(line.split('\n')[0].split(', ')[2].split('FP: ')[1]) for line in lines if 'Tversky' in line]
    FN = [float(line.split('\n')[0].split(', ')[3].split('FN: ')[1]) for line in lines if 'Tversky' in line]
    
    #F1 = 2TP/(2TP+FP+FN)
    TP2 = [i * 2 for i in TP]
    denom = list( map(add, TP2, FP) )
    denom_ = list( map(add, denom, FN) )
    F1 = [i / j for i, j in zip(TP2, denom_)]
    
    return F1
```

```python
def getF1_BCE(fn='Output_BCE.txt', norm_flag=1):
    with open(fn) as f:
        lines = f.readlines()
   
    if norm_flag == 1:
        lines_ = []
        for i in range(len(lines)):
            if 'Dice' in lines[i]:
                lines_.append(lines[i+1])
        lines = lines_

    temp_list = [line.split('\n')[0].split(', ') for line in lines]
    TP = [float(l[0].split('TP: ')[1]) for l in temp_list]
    FP = [float(l[1].split('FP: ')[1]) for l in temp_list]
    FN = [float(l[2].split('FN: ')[1]) for l in temp_list]
    
    #F1 = 2TP/(2TP+FP+FN)
    TP2 = [i * 2 for i in TP]
    denom = list( map(add, TP2, FP) )
    denom_ = list( map(add, denom, FN) )
    F1 = [i / j for i, j in zip(TP2, denom_)]
    
    return F1
```
